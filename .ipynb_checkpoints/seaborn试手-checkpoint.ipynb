```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
import pandas as pd
#读取原始数据，分离Y
train_file=r'C:\Python\kaggle\Titanic\data\train.csv'
test_file=r'C:\Python\kaggle\Titanic\data\test.csv'
train_data=pd.read_csv(train_file).set_index(['PassengerId'])
test_data=pd.read_csv(test_file).set_index(['PassengerId'])

survived=train_data['Survived']
train_data_original=train_data
train_data=train_data.drop(['Survived'],axis=1)

data=pd.concat([train_data,test_data])
data=data.drop(['Name','Cabin','Ticket'],axis=1)
```
