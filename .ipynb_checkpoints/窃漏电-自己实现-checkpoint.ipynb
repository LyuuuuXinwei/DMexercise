```python
import numpy as np
import pandas as pd
import os
from sklearn.tree import DecisionTreeClassifier
```

```python
#读文件预备动作
data_file=r'C:\Python\DM exercise\dm\chapter6\demo\data\model.xls'
o_data=pd.read_excel(data_file)
```

```python
o_data
```

```python
n=14
x=o_data.iloc[:,:3].as_matrix()
y=o_data.iloc[:,3].as_matrix()

from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=n)

dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
from sklearn.externals import joblib
decision_tree_file=r'C:\Python\DM exercise\dm\chapter6\demo\data\tree.pkl'
joblib.dump(dt,decision_tree_file)
```

```python
y_pre=dt.predict(x_test)
y_pre
```

```python
y_test
```

```python
np.stack((y_pre,y_test)).T
```

```python
pre_n_real=pd.DataFrame(np.stack((y_pre,y_test)).T,columns=[u'预测',u'真实'])
```

```python
pre_n_real
```

```python
matrix=pd.DataFrame(np.zeros([2,2],dtype=int))
matrix.index.name=u'预测'
matrix.columns.name=u'实际'
matrix
```

```python
pre_n_real[np.array(pre_n_real[u'预测']==0) & np.array(pre_n_real[u'真实']==0)]
```

```python
for i in range(2):
    for j in range(2):
        matrix.iloc[i,j]=len(pre_n_real[np.array(pre_n_real[u'预测']==i) & np.array(pre_n_real[u'真实']==j)])
```
