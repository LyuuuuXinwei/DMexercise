```python
import numpy as np
import pandas as pd
import os

data_file=r'C:\Python\DM exercise\dm\chapter5\demo\data\consumption_data.xls'
o_data=pd.read_excel(data_file)
o_data=o_data.drop(['Id'],axis=1)
o_data_zs=1.0*(o_data-o_data.mean())/o_data.std()

k=3
iteration=500

from sklearn.cluster import KMeans
model=KMeans(n_clusters=k,n_jobs=4,max_iter=iteration)
model.fit(o_data_zs)

data_clustered=pd.concat([o_data,pd.Series(model.labels_,index=o_data.index)],axis=1)
data_clustered.columns=list(o_data.columns)+['class']
```

```python
data_clustered
```

```python
# data_clustered=data_clustered.drop(['Id'],axis=1)
```

```python
data_in_each_class=data_clustered[data_clustered['class']==1].iloc[:,:3]
data_in_each_class
```

```python
distance=[]
for i in range(k):
    data_in_each_class=data_clustered[data_clustered['class']==i].iloc[:,:3]
    dis_temp=data_in_each_class - model.cluster_centers_[i]
    dis=dis_temp.apply(np.linalg.norm,axis=1)
    relative_dis=dis/dis.median()
    print('在%s类别中，相对距离输出为:%s' % (i,relative_dis))
    print('--'*10)
    distance.append(relative_dis)
```

```python
distance
```

```python
distance=pd.concat(distance) #TODO:列表和series这样互相转化？？？series被append然后又可以concat成series
```
