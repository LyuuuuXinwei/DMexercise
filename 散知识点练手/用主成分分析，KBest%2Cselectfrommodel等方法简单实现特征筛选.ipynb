```python
#相关性分析，相关系数越靠近0越不相关，并不能用此判断，原因是故障类别的编号无意义，而计量数据可以
data.corr(method='pearson')[u'故障类别']
```

```python
#主成分分析，方差跟整体方差没有多大差距，这样的特征也被称为主成分，本质计算特征之间的相关性，适用于连续属性,只需要X
from sklearn.decomposition import PCA
n=7
x=data.iloc[:,:7]
pca=PCA(n_components=n)
xd=pca.fit_transform(x)

pca.explained_variance_ratio_
```

```python
#以下代码是被转换过的主成分，所以除了图像识别等，有解释性需求的筛选最好别用
n=5
x=data.iloc[:,:7]
pca=PCA(n_components=n)
xd=pca.fit_transform(x)
# low_3=pd.DataFrame(xd)
low_3=pca.inverse_transform(xd)
low_3
```

```python
#KBest x,y都需要
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectPercentile
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import mutual_info_classif

x=data.iloc[:,:7]
y=data.iloc[:,7]
```

```python
'''
卡方检验就是统计样本的实际观测值与理论推断值之间的偏离程度
卡方值越大，越不符合；卡方值越小，偏差越小，越趋于符合，若两个值完全相等时，卡方值就为0
返回：特征统计和特征P值 
可以用于判断分类变量的得分
'''
chi2(x,y) 
```

```python
a=chi2(x,y)[1]
a=list(a)
b=sorted(a)
print(a)
print(b)#按照p值，特征重要度排序为：4-2-5-6-1，其中4显著其余雷同
```

```python
f_classif(x,y) #选F值大或P小，肉眼看排序，特征4显著地大
'''
其通常是用来分析用了超过一个参数的统计模型，以判断该模型中的全部或一部分参数是否适合用来估计母体
返回：特征F值，特征P值
可以用于判断分类变量的得分
'''
```
