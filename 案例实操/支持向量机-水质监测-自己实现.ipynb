```python
import numpy as np
import pandas as pd
import os
from sklearn import svm
```

```python
data_file=r'C:\Python\DM exercise\dm\chapter9\demo\data\moment.csv'
o_data=pd.read_csv(data_file,encoding='gbk')
```

```python
o_data
```

```python
n=14
x=o_data.iloc[:,2:].as_matrix()*30
y=o_data.iloc[:,0].as_matrix()
```

```python
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=n)
```

```python
model=svm.SVC()
model.fit(x_train,y_train)
```

```python
import pickle
pickle.dump(model,open(r'C:\Python\DM exercise\dm\chapter9\demo\data\svm.model','wb'))
```

```python
from sklearn import metrics
m_train=metrics.confusion_matrix(y_train,model.predict(x_train))
m_test=metrics.confusion_matrix(y_test,model.predict(x_test))
```
